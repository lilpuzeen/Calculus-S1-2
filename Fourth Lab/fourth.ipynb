### Функция:
$${\normalsize f(x, y) = x^2\,y^2\,(5 - 2x - 4y) = 5x^2 y^2 - 2x^3 y^2 - 4x^2 y^3}$$
<hr>

<p style="font-weight: bold; font-size: 24px;">Аналитический метод</p>

<p style="font-weight: bold; font-size: 18px;">1. Посчитаем частные производные</p>

$${\normalsize \frac{\partial f}{\partial x} = 10xy^2 - 6x^2 y^2 - 8xy^3 = 2xy^2 (5 - 3x - 4y)}$$
<br>
$${\normalsize \frac{\partial f}{\partial y} = 10x^2 y - 4x^3 y - 12x^2 y^2 = 2x^2 y (5 - 2x - 6y)}$$

<p style="font-weight: bold; font-size: 18px;">Найдем стационарные точки</p>

$$
{\normalsize
\begin{equation}
    \begin{cases}
      2xy^2\,(5 - 3x - 4y) = 0\\
      \\
      2x^2 y\,(5 - 2x - 6y) = 0
    \end{cases}\,
\end{equation}
}
$$

<p style="font-weight: bold; font-size: 18px;">Решения системы</p>

$${\normalsize M = \left\{\left(1, \frac{1}{2}\right), \left(t, 0\right), \left(0, t\right)\right\}, \quad t \in \mathbb{R}}$$

<p style="font-weight: bold; font-size: 14px;">Пояснение: если одна из переменных равняется нулю, то другая не равная нулю может принимать любое значение</p>
<hr>

<p style="font-weight: bold; font-size: 18px;">2. Построим матрицу Гессе</p>

<p style="font-weight: bold; font-size: 18px;">Матрица Гессе имеет вид:</p>

$$
{\normalsize H(\,f\,) =
\begin{pmatrix}
{\frac{\partial^2 f}{\partial x^2}} & {\frac{\partial^2 f}{\partial x \partial y}}\\
{\frac{\partial^2 f}{\partial y\partial x}} & {\frac{\partial^2 f}{\partial y^2}}
\end{pmatrix}
}
$$

$${\normalsize \frac{\partial^2 f}{\partial x^2} = 10y^2 - 12xy^2 - 8y^3}$$
<br>
$${\normalsize \frac{\partial^2 f}{\partial y^2} = 10x^2 - 4x^3 - 24x^2 y}$$
<br>
$${\normalsize \frac{\partial^2 f}{\partial x \partial y} = 20xy - 12x^2 y - 24xy^2 = \frac{\partial^2 f}{\partial y \partial x}}$$

<p style="font-weight: bold; font-size: 18px;">Получим матрицу:</p>

$$
{\normalsize H(\,f\,) = 
\begin{pmatrix}
{\normalsize 10y^2 - 12xy^2 - 8y^3} & {\normalsize 20xy - 12x^2 y - 24xy^2}\\
{\normalsize 20xy - 12x^2 y - 24xy^2} & {\normalsize 10x^2 - 4x^3 - 24x^2 y}
\end{pmatrix}
}
$$

<p style="font-weight: bold; font-size: 18px;">Подставим в матрицу значения полученных ранее точек</p>

$$
{\normalsize M_1 = \left(1, \frac{1}{2}\right)}
$$

$$
{\normalsize H_{M_1}(\,f\,) = 
\begin{pmatrix}
{\normalsize -1.5} & {\normalsize -2}\\
{\normalsize -2} & {\normalsize -6}
\end{pmatrix}
}
$$
<br>

$$
{\normalsize M_2 = \left(0, t\right)}
$$

$$
{\normalsize H_{M_2}(\,f\,) = 
\begin{pmatrix}
{\normalsize 10t^2 - 8t} & {\normalsize 0}\\
{\normalsize 0} & {\normalsize 0}
\end{pmatrix}
}
$$
<br>

$$
{\normalsize M_3 = \left(t, 0\right)}
$$

$$
{\normalsize H_{M_3}(\,f\,) = 
\begin{pmatrix}
{\normalsize 0} & {\normalsize 0}\\
{\normalsize 0} & {\normalsize 10t^2 - 4t^3}
\end{pmatrix}
}
$$

<p style="font-weight: bold; font-size: 18px;">Найдем гессианы</p>

$$
{\normalsize \det(\normalsize H_{M_1}) = 5}
$$
<br>
$$
{\normalsize \det(\normalsize H_{M_2}) = 0}
$$
<br>
$$
{\normalsize \det(\normalsize H_{M_3}) = 0}
$$

<p style="text-align: center; font-size: 18px;">Исходя из критерия Сильвестра для квадратичной формы $H_{M_1}(\,f\,),$ мы можем сказать что она отрицательно определенная: возьмем с минусом $H_{M_1}(\,f\,),$ и получим положительно определенную</p>

<p style="text-align: center; font-size: 18px;">Следовательно, имея что $\det(\normalsize H_{M_1}) = 5 > 0$, точка $M_1$ - max</p>

$${\normalsize \text{Значение функции - } f(M_1) = 0.25}$$
<hr>

<p style="font-weight: bold; font-size: 18px;">3. Заметим что для точек $M_2, M_3$ не выполняется достаточное условие. Проверим их дополнительно</p>


$${\normalsize \text{Для } M_2 = (0, t)}$$

$${\normalsize f(M_2) = 0}$$
<br>
$${\normalsize f(M_2 + \Delta) = f(0 + \Delta x, t + \Delta y) = (\Delta x)^2 (t + \Delta y)^2 (5 - 2\Delta x - 4t - 4\Delta y)}$$
<br>
$${\normalsize \text{Заметим } (\Delta x)^2 \ge 0 \text{ и } (t + \Delta y)^2 \ge 0}$$
<br>
$${\normalsize с - (2\Delta x + 4\Delta y), \, c - \text{ константа}}$$
<br>
$${\normalsize \text{Понятно что в зависимости от значений } \Delta x \text{ и } \Delta y \rightarrow (2\Delta x + 4\Delta y) \text{ будет иметь разный знак} \Rightarrow}$$
<br>
$${\normalsize \text{Иногда } f(M_2 + \Delta) > 0 \quad \text{А иногда, } f(M_2 + \Delta) < 0 \quad \Rightarrow}$$
<br>
$${\normalsize M_2 - \text{ не экстремум}}$$
<br>
$${\normalsize M_3 - \text{ по точно таким же рассуждениям, не экстремум}}$$

<p style="font-weight: bold; font-size: 24px;">Численный метод</p>

```python
# Импортируем все нужные библиотеки (очень жаль что нельзя использовать autograd)
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# Задаем изначальную функцию
def f(x, y):
    return 5 * x**2 * y**2 - 2 * x**3 * y**2 - 4 * x**2 * y**3
```

```python
# Задаем градиент функции
def grad(x, y):
    partial_x = 2 * x * y**2 * (5 - 3*x - 4*y)
    partial_y = 2* x**2 * y * (5 - 2*x - 6*y)
    return -partial_x, -partial_y
```

```python
x_start, y_start = 1, 1  # Стартовое приближение
COEF = 0.1  # Коэффицент  
ITERATIONS = 100 # Число итераций
EPS = 1e-6  # Эпсилон для условия остановки алгоритма
```

```python
x_coord, y_coord, z_coord = [], [], []
```

```python
# Градиентный спуск
%time
for i in range(ITERATIONS):
    x_i, y_i = x_start - COEF * grad(x_start, y_start)[0], y_start - COEF * grad(x_start, y_start)[1]
    f_i = f(x_i, y_i)
    delta = abs(f_i - f(x_start, y_start))
    
    x_coord.append(x_i)
    y_coord.append(y_i)
    z_coord.append(f_i)

    if delta < EPS:
        last_iter_count = i
        break
    x_start, y_start = x_i, y_i
```

```python
print(f"Критерий остановки: abs(Delta f) < {EPS}")
print(f"Число итераций: {last_iter_count}")
print(f"Полученная точка: ({x_i}, {y_i})")
print(f"Полученное значение функции: {f_i}")
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(1, 2, 1, projection='3d')

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax.plot_surface(X, Y, Z, cmap='plasma')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Function Graph')

ax = fig.add_subplot(1, 2, 2)
ax.contour(X, Y, Z, 50, cmap='plasma')
ax.scatter(x_coord, y_coord, color='red', s=10)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Level set')

plt.show()
```
