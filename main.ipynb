### Последовательность:
$$x_n = (1 + \sin{\frac{\pi n}{2}}) \cdot \frac{n - 3}{n + 5}$$

<p style="font-weight: bold; font-size: 14px;">2. Найдем значения: $\sup{x_n}, \inf{x_n}, \overline{\lim}{x_n}, \underline{\lim}{x_n}$</p>

<p style="font-weight: bold; font-size: 14px;">3. Исследуем на наличие: $\max{x_n}, \, \min{x_n}$</p>

<h2 style="text-align:center">Численный Метод</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("seaborn")
```

```python
def function(n):
    return (1 + np.sin((np.pi * n) / 2)) * ((n - 3) / (n + 5))
```

```python
x = np.arange(-1000, 1000)
y = function(x)
y1 = np.sin(x)
```

```python
plt.xlim(-100, 100)
plt.ylim(-2, 2)
plt.plot(x, y)
plt.show()
```
