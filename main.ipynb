### Последовательность:
$${\normalsize x_n = (1 + \sin{\frac{\pi n}{2}}) \cdot \frac{n - 3}{n + 5}}$$

<h2 style="text-align:center">Аналитический Метод</h2>

<p style="font-weight: bold; font-size: 18px;">1. Исследуем последовательность $x_n$ на сходимость</p>

<p style="font-size: 16px; text-align: center; font-weight: bold"> Для начала посчитаем первые 6 членов последовательности </p>

<table style="border: 1px solid">
<tr>
    <td style="text-align:left">${\normalsize x_1 = (1 + \sin(\frac{\pi}{2})) \cdot \frac{1 - 3}{1 + 5} = \frac{-2}{3}}$</td>
    <td style="text-align:left">${\normalsize x_2 = (1 + \sin(\frac{2 \pi}{2})) \cdot \frac{2 - 3}{2 + 5} = \frac{-1}{7}}$</td>
</tr>
<tr>
    <td style="text-align:left">${\normalsize x_3 = (1 + \sin(\frac{3 \pi}{2})) \cdot \frac{3 - 3}{3 + 5} = 0}$</td>
    <td style="text-align:left">${\normalsize x_4 = (1 + \sin(\frac{4 \pi}{2})) \cdot \frac{4 - 3}{4 + 5} = \frac{1}{9}}$</td>
</tr>
<tr>
    <td style="text-align:left">${\normalsize x_5 = (1 + \sin(\frac{5 \pi}{2})) \cdot \frac{5 - 3}{5 + 5} = \frac{4}{10}}$</td>
    <td style="text-align:left">${\normalsize x_6 = (1 + \sin(\frac{6 \pi}{2})) \cdot \frac{6 - 3}{6 + 5} = \frac{3}{11}}$</td>
</tr>

<p style="font-size: 16px; text-align: center; font-weight:bold"><em>Синус - функция периодическая</em>, потому выделим следующие подпоследовательности и найдем частичные пределы</p>

<p style="font-size: 15px">Если $n = 2k; k \in \mathbb{N}$ то:</p>
$${\normalsize x_n = x_{2k} = (1 + \sin(2\pi k)) \cdot \frac{2k - 3}{2k + 5} = \frac{2k - 3}{2k + 5}}$$
<p style="font-size: 10px; text-align:right">Так как $\sin(2\pi k)$ всегда равен 0</p>
<br>
<p style="font-size: 15px">Найдем частичный предел:</p>
$${\normalsize \lim_{k \rightarrow \infty}{\frac{2k - 3}{2k + 5}} = 1}$$
<hr>

<p style="font-size: 15px">Если $n = 4k - 1; k \in \mathbb{N}$ то:</p>
$${\normalsize x_n = x_{4k - 1} = \left(1 + \sin\left(\frac{(4k - 1) \pi}{2}\right)\right) \cdot \frac{4k - 4}{4k + 4} = 0}$$
<p style="font-size: 10px; text-align:right">Так как $\sin\left(\frac{(4k -1) \pi)}{2}\right)$ всегда равен -1</p>
<br>
<p style="font-size: 15px">Найдем частичный предел:</p>
$${\normalsize \lim_{k \rightarrow \infty} 0 = 0}$$
<hr>

<p style="font-size: 15px">Если $n = 4k - 3; k \in \mathbb{N}$ то:</p>
$${\normalsize x_n = x_{4k - 3} = \left(1 + \sin\left(\frac{(4k - 3) \pi)}{2}\right)\right) \cdot \frac{4k - 6}{4k + 2} = \frac{8k - 12}{4k + 2}}$$
<p style="font-size: 10px; text-align:right">Так как $\sin\left(\frac{(4k - 3) \pi)}{2}\right)$ всегда равен 1</p>
<br>
<p style="font-size: 15px">Найдем частичный предел:</p>
$${\normalsize \lim_{k \rightarrow \infty}{\frac{8k - 12}{4k + 2}} = 2}$$
<hr>

<p style="font-size: 15px">В итоге получили множество частичных пределов:</p>
<p style="font-size: 18px">$${\normalsize L = \{0, 1, 2\}}$$</p>
<p style="font-size: 15px; text-align: center">Так как ${\normalsize \overline{\lim}{x_n} \ne \underline{\lim}{x_n}}$, наша <strong>последовательность $x_n$ расходится</strong></p>

<p style="font-size: 18px; text-align:center">Найдем верхний и нижний предел последовательности</p>
<p style="font-size: 15px"><strong>Верхний предел последовательности</strong> - это наибольший предел из множества частичных пределов</p>
<p style="font-size: 15px"><strong>Нижний предел последовательности</strong> - это наименьший предел из множества частичных пределов</p>


<p style="font-size: 20px">$${\overline{\lim}{x_n} = 2 \qquad \underline{\lim}{x_n} = 0}$$</p>
<hr>

<p style="font-weight: bold; font-size: 18px;">2. Найдем значения: ${\sup{x_n}, \inf{x_n}, \overline{\lim}{x_n}, \underline{\lim}{x_n}}$</p>

<p style="font-size:16px; text-align:center">Значения ${\normalsize \overline{\lim}{x_n}}$ и ${\normalsize \underline{\lim}{x_n}}$ мы нашли уже пунктом выше</p>

<p style="font-size: 16px;text-align:center; font-weight: bold">Найдем ${\normalsize \inf{x_n}}$ и ${\normalsize \sup{x_n}}$</p>

<p style="font-weight: bold; font-size: 16px;">2.1. ${\normalsize \inf{x_n}}$</p>

<p style="font-size:15px">В нашей последовательности есть два множителя. Оценим их сверху и снизу, исследуем какие значения могут принимать</p>
<br>
$${\normalsize (1 + \sin{\frac{\pi n}{2}}) \in \{0, 1, 2\};\,n \in \mathbb{N}}\\$$
$${\normalsize \frac{n - 3}{n + 5} \in \left[-\frac{1}{3}; 1\right);\,n \in \mathbb{N}}$$

<p style="font-size:15px">Исходя из найденного, найдем наименьшее значение произведения этих двух множителей:</p>
<br>
$${\normalsize \left(1 + \sin{\frac{\pi n}{2}}\right) \cdot \frac{n - 3}{n + 5} = -\frac{2}{3}}$$

<p style="font-size:15px">Однако заметим, что найденное выше наименьшее значение достигается при $n = 1$:</p>
<br>
$${\normalsize \left(1 + \sin{\frac{\pi}{2}}\right) \cdot \frac{1 - 3}{1 + 5} = -\frac{2}{3}}$$

<p style="font-size:15px">Следовательно получаем что:</p>
<br>
$${\normalsize \inf{x_n} = -\frac{2}{3}}$$
<hr>

<p style="font-weight: bold; font-size: 16px;">2.2. ${\normalsize \sup{x_n}}$</p>

<p style="font-size:15px">Исследуем еще раз оба множителя на предмет принимаемых ими значений:</p>
<br>
$${\normalsize \left(1 + \sin{\frac{\pi n}{2}}\right) \in \{0, 1, 2\};\,n \in \mathbb{N}}\\$$
$${\normalsize \frac{n - 3}{n + 5} \in \left[-\frac{1}{3}; 1\right);\,n \in \mathbb{N}}$$

<p style="font-size:15px">Найдем наибольшее значение их произведения и предположим что оно будет являться ${\normalsize \sup{x_n}}$</p>
<br>
$${\normalsize \left(1 + \sin{\frac{\pi n}{2}}\right) \cdot \frac{n - 3}{n + 5} \approx 2}\\$$
<p style="font-size:15px">Но! Данное произведение никогда не достигнет значения <em>2</em>, так как у второго множителя конкретного максимума не существует</p>
<hr>

<p style="font-size:15px; text-align:center">Докажем что ${\normalsize \sup{x_n} = 2}$. Для этого проверим его по критерию супремума</p>

<p style="font-size:15px">1. Предполагаемый супремум $S$ должен являться верхней границей:</p>
<br>
$${\normalsize \forall x \in {x_n}: x \le S}$$
<p style="font-size:15px">Очевидно что является верхней границей, так как мы взяли наибольшее значение произведения множителей в формуле последовательности</p>

<p style="font-size:15px">2. Предполагаемый супремум $S$ должен удовлетворять следующему условию: </p>
<br>
$${\normalsize \forall S' \lt S, \, \exists x \in {x_n}: S' \lt x}$$
<p style="font-size: 16px; text-align:center;">Докажем</p>

<p>$${\large \forall \varepsilon > 0,\, \exists n_0: x_n \gt 2 - \varepsilon}$$</p>

<p style="text-align:left">${\normalsize \text{Пусть } n = 2k, \, k \in \mathbb{N}:}$</p>

<p>$${\normalsize \frac{2k - 3}{2k + 5} > 2 - \varepsilon}$$</p>

<p>$${\normalsize \frac{2k + 13}{2k + 5} < \varepsilon}$$</p>

<p>$${\normalsize \frac{2k + 13}{2k + 5} < \frac{2k + 13}{2k} < \varepsilon}$$</p>

$${\normalsize 2k + 13 < 2k\varepsilon}$$

$${\normalsize 13 < 2k\varepsilon - 2k}$$

$${\normalsize 13 < 2k (\varepsilon - 1)}$$

$${\normalsize n_0 = \left\lfloor\frac{13}{2(\varepsilon - 1)} \right\rfloor + 1}$$

<p style="font-size:16px">Данная цепочка рассуждений нас приводит к тому, что мы всегда сможем взять такое число $n = 2k$, чтобы оно удовлетворяло критерию супремума.</p>

<p style="font-size:16px; font-weight:bold; text-align: center;">Значит действительно:</p>
<br>
$${\normalsize \sup{x_n} = 2}$$
<hr>

<p style="font-weight: bold; font-size: 18px;">3. Исследуем на наличие: $\max{x_n}, \, \min{x_n}$</p>

<p style="font-size: 15px">Максимального элемента нет, так как наша последовательность хоть и ограничена сверху, но она не достигает своего супремума.</p>
<p style="font-size: 15px">Минимальный элемент есть, так как у нас существует конкретный достигаемый инфимум</p>
<br>
$${\normalsize \min{x_n} = \inf{x_n} = -\frac{2}{3}}$$

<p style="font-weight: bold; font-size: 18px;">4. Определение предела для подпоследовательности</p>

<p style="font-size: 15px">Выберем подпоследовательность последовательности ${\normalsize x_n}$ при ${\normalsize n = 4k - 1}$ и обозначим ее ${\normalsize y_k}$</p>

$${\normalsize y_k = 0}$$

<p>$${\normalsize \forall \varepsilon > 0,\, \exists k_0 \in \mathbb{N}: \forall k \ge k_0 \Rightarrow |y_k - 0| < \varepsilon}$$</p>

<p style="font-size: 15px">Выбранная мною подпоследовательность является стационарной, всегда равная 0</p>

$${\normalsize |y_k - 0| < \varepsilon}$$

$${\normalsize |0 - 0| < \varepsilon}$$

$${\normalsize 0 < \varepsilon}$$

<p style="font-size:16px; text-align:center;">Следовательно, можно взять ${\normalsize k_0}$ за любой номер</p>
<hr>

<h2 style="text-align:center">Численный Метод</h2>

<p style="font-weight: bold; font-size: 18px;">1. Построим график последовательности $x_n$</p>

```python
import numpy as np  # Библиотека для работы с массивами чисел
import matplotlib.pyplot as plt  # Библиотека для построения графиков
%matplotlib inline
```

```python
plt.style.use(["science", "notebook", "grid", "high-vis"])  # Стиль координатной плоскости
```

<p style="text-align:center; font-weight:bold">Зададим нашу последовательность в виде функции</p>

```python
def function(n):
    return (1 + np.sin((np.pi * n) / 2)) * ((n - 3) / (n + 5))
```

```python
X = np.arange(1, 100 + 5)  # Отметим первые 100 точек на оси абсцисс
X_half = np.arange(1, 50+3)  # Для подпоследовательности
Y = function(X)  # Применим функцию на массив Х, тем самым получим значения в точках
```

```python
# Задание 2.4

eps = 0.01  # Эпсилон
x_m = 0  # Переменная в которую положим значение которое окажется больше чем sup - eps
sup = 2  #  Супремум
m = 0  #  Номер этого значения
for i in range(1, 1_000_000):
    tmp = function(i)
    if tmp > sup - eps:  #  Критерий супремума
        x_m = tmp
        m = i
        print(f"Найденное значение большее супремума: {x_m}")
        print(f"Номер: {i}")
        break

X_sup = np.arange(m - 100, m + 1)
Y_sup = function(X_sup)
```

```python
plt.figure(figsize=(20, 10))  # Зададим размер окна с коорд. плоскостью
plt.axvline(x=0, lw=2, color="black")  # Построим х=0
plt.axhline(y=0, lw=1, color="black")  # Построим у=0
plt.xlabel("$n$", fontsize=20)  # Подпишем ось абсцисс
plt.ylabel("$x_n$", fontsize=24)  # Подпишем ось ординат

# Построим график последовательности
plt.plot(X, Y, "o--", lw=2, ms=9, label="Main Sequence")

# Прямая линия, отвечает за супремум и верхний предел 
plt.plot(X, X*0 + 2, ":", lw=4, label="Supremum and Top Limit")

# Прямая линия, отвечает за нижний предел
plt.plot(X, X*0, "--", lw=4, color="orange", label="Bottom Limit")

# Прямая линия, отвечает за инфимум
plt.plot(X, X*0 + function(X[0]), lw=4, label="Infimum")

# Подпоследовательность при n = 2k
plt.plot(2*X_half, (2*X_half - 3) / (2*X_half + 5), "o", lw=4, ms=12, label="Subsequence: n = 2k")

# Точка из задания 2.4
plt.plot(100, x_m, "o", ms=20, label="Checkpoint for Supremum")
plt.text(80, 2, f"$m = {m}, f(x_m) = {x_m}$", fontsize=8, bbox=dict(facecolor="white", edgecolor="black"))

plt.legend(loc="center left", bbox_to_anchor=(1, 0.5), edgecolor="black")  # Отобразим легенду графика
plt.show()
```

<p><small>В легенде указаны цвета тех или иных значений</small></p>

<p style="font-weight: bold; font-size: 18px;">3. По данному $\varepsilon > 0$ найдем $n_0$. Отметим значение предела для данной подпоследовательности построенной начиная с $n_0$</p>

<p style="font-size:15px">Для той же выбранной последовательности ${\normalsize n = 2k}$ нарисуем ее график, и четко изобразим её предел начиная с некоторого номера.</p>

<p>Если ${\normalsize n = 2k}$, то:</p>
<br>
$${\normalsize x_n = x_{2k} = \frac{2k - 3}{2k + 5}}$$

<p style="font-size:15px">Определение предела: </p>
<br>
<p>$${\normalsize \forall \varepsilon > 0,\, \exists k_0 \in \mathbb{N}: \forall k \ge k_0 \Rightarrow |x_{2k} - 1| < \varepsilon}$$</p>

$${\normalsize \left|\frac{2k - 3}{2k + 5} - 1\right| < \varepsilon}$$

$${\normalsize \left|\frac{-8}{2k + 5}\right| < \varepsilon}$$

$${\normalsize \frac{8}{2k + 5} < \frac{8}{2k} = \frac{4}{k} < \varepsilon}$$

$${\normalsize k_0 = \left\lfloor \frac{4}{\varepsilon} \right\rfloor + 1}$$

```python
# Функция для получения информации по эпсилон-окрестности, по заданному эпсилон
def epsilon_info(eps: float) -> tuple:
    LIMIT = 1  # Предел подпоследовательности
    n_0 = np.floor(4 / eps) + 1  # Номер, с которого график попадает в эпс-окрестность
    eps_delta_bot = LIMIT - eps  # Линия ограничивающая эпс-окрестность: нижняя
    eps_delta_top = LIMIT + eps  # Линия ограничивающая эпс-окрестность: верхняя
    X_sub = np.array([2*k for k in range(int(n_0), int(n_0) + 100)])
    Y_sub = function(X_sub)
    
    return (eps_delta_bot, eps_delta_top, X_sub, Y_sub)
```

```python
# Инициализируем три графика 
fig, axes = plt.subplots(3, 1, figsize=(15, 15))

# Аналогичные действия для всех графиков
# График эпсилон = 0.01
ax1 = axes[0]
ax1.set_title(r"$\varepsilon = 0.01$ Plot", fontsize=20)  # Название у графика
# Из функции подтягиваем все нужные значения
ax1_bot_limit, ax1_top_limit, ax1X, ax1Y = epsilon_info(0.01)
ax1.set_xlabel("k", fontsize=20)  # Лэйбл по оси Ох
ax1.set_ylabel(r"$x_{2k}$", fontsize=20)  # Лэйбл по оси Оу
ax1.plot(ax1X, ax1Y, "-", lw=4, label="Subsequence")  # График подпоследовательнсоти

# Верхняя граница эпсилон окрестности
ax1.plot(ax1X, ax1X*0 + ax1_top_limit, ":", lw=4, color="magenta", label="Epsilon-Delta upper bound")

# Нижняя граница эпсилон окрестности
ax1.plot(ax1X, ax1X*0 + ax1_bot_limit, ":", lw=4, color="indigo", label="Epsilon-Delta lower bound")

# Предел подпоследовательности
ax1.plot(ax1X, ax1X*0 + 1, "--", lw=4, color="red", label="Subsequence Limit")

# Текст
ax1.text(805, 1.0075, 
         r"График подпоследовательности входит в $\varepsilon$-окрестность начиная с номера 800", 
         fontsize=14,bbox=dict(facecolor="white", edgecolor="black"))

# Заливка эпсилон-окрестности
ax1.fill_between(ax1X, ax1_bot_limit, ax1_top_limit, color="yellow", alpha=0.5)

# Легенда
ax1.legend(loc="center left", bbox_to_anchor=(1, 0.5), edgecolor="black")


# Аналогичные действия для остальных графиков
# График эпсилон = 0.001
ax2 = axes[1]
ax2.set_title(r"$\varepsilon = 0.001$ Plot", fontsize=20)
ax2_bot_limit, ax2_top_limit, ax2X, ax2Y = epsilon_info(0.001)
ax2.set_xlabel("k", fontsize=20)
ax2.set_ylabel(r"$x_{2k}$", fontsize=20)
ax2.plot(ax2X, ax2Y, "o", lw=4, ms=6, label="Subsequence")
ax2.plot(ax2X, ax2X*0 + ax2_top_limit, ":", lw=4, color="magenta", label="Epsilon-Delta upper bound")
ax2.plot(ax2X, ax2X*0 + ax2_bot_limit, "-", lw=2, color="indigo", label="Epsilon-Delta lower bound")
ax2.plot(ax2X, ax2X*0 + 1, "-", lw=4, color="red", label="Subsequence Limit")
ax2.text(8005, 1.0005, 
         r"График подпоследовательности входит в $\varepsilon$-окрестность начиная с номера 8000", 
         fontsize=14,bbox=dict(facecolor="white", edgecolor="black"))
ax2.fill_between(ax2X, ax2_bot_limit, ax2_top_limit, color="yellow", alpha=0.5)
ax2.set_ylim(0.99900 - 0.0009, 0.99900 + 0.002)
ax2.legend(loc="center left", bbox_to_anchor=(1, 0.5), edgecolor="black")

# График эпсилон = 0.0001
ax3 = axes[2]
ax3.set_title(r"$\varepsilon = 0.0001$ Plot", fontsize=20)
ax3_bot_limit, ax3_top_limit, ax3X, ax3Y = epsilon_info(0.0001)
ax3.set_xlabel("k", fontsize=20)
ax3.set_ylabel(r"$x_{2k}$", fontsize=20)
ax3.plot(ax3X, ax3Y, "o", lw=6, label="Subsequence")
ax3.plot(ax3X, ax3X*0 + ax3_top_limit, ":", lw=4, color="magenta", label="Epsilon-Delta upper bound")
ax3.plot(ax3X, ax3X*0 + ax3_bot_limit, "-", lw=2, color="indigo", label="Epsilon-Delta lower bound")
ax3.plot(ax3X, ax3X*0 + 1, "--", lw=4, color="red", label="Subsequence Limit")
ax3.text(80010, 0.99995, 
         r"График подпоследовательности входит в $\varepsilon$-окрестность начиная с номера 80000", 
         fontsize=14, bbox=dict(facecolor="white", edgecolor="black"))
ax3.fill_between(ax3X, ax3_bot_limit, ax3_top_limit, color="yellow", alpha=0.5)
ax3.axis([ax3X[0], ax3X[-1], 0.9999 - 0.0001, 0.9999 + 0.0001])
ax3.legend(loc="center left", bbox_to_anchor=(1, 0.5), edgecolor="black")

fig.tight_layout()
plt.show()
```
