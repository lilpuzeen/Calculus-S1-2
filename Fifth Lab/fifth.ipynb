# Интеграл:
$${\large \int_L{x^2 + y^2}dx + 3xydy}$$
### где
$${\large L - \text{граница полукруга } D: x^2 + y^2 \le 1, x \ge y, \text{ проходимая по часовой стрелке}}$$

# Нарисуем кривую интегрирования

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy.solvers import solve
from sympy import Symbol
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем сетку координат
x = np.linspace(-2, 2, 1000)
y = np.linspace(-2, 2, 1000)
X, Y = np.meshgrid(x, y)

# Условия
condition1 = X**2 + Y**2 <= 1
condition2 = X >= Y

# Объединяем условия
conditions = condition1 & condition2

# Построение графика
plt.figure(figsize=(7, 8))
plt.scatter(X[conditions], Y[conditions], color='blue', s=1)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

plt.xlabel('x')
plt.ylabel('y')
plt.title('График на неравенствах x^2 + y^2 ≤ 1 и x ≥ y')
plt.grid(True)
plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))
plt.show()
```
