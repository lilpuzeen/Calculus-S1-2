```python
import numpy as np  # Библиотека для работы с массивами чисел
import matplotlib.pyplot as plt  # Библиотека для построения графиков
%matplotlib inline
```

```python
plt.style.use(["science", "notebook", "grid", "high-vis"])  # Стиль координатной плоскости
```

```python
def function(x):
    return np.cos((np.pi / 6) + x)
```

```python
def get_n_derivative(n):
    return np.cos((np.pi / 6) + ((np.pi * n) / 2))
```

```python
def taylor_polynom(x, n):
    result_sum = 0
    for k in range(n + 1):
        result_sum += (1 / np.math.factorial(k)) * get_n_derivative(k) * np.power(x, k)
    return result_sum
```

```python
X = np.arange(-5, 1 + 5)
Y = function(X)
Y_1 = taylor_polynom(X, 1)
Y_2 = taylor_polynom(X, 2)
Y_9 = taylor_polynom(X, 9)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(15, 20))

# Аналогичные действия для всех графиков

ax1 = axes[0]
ax1.set_title("Taylor polynom for n in [1, 2, 3]", fontsize=20)  # Название у графика
# Из функции подтягиваем все нужные значения
ax1.set_xlabel("x", fontsize=20)  # Лэйбл по оси Ох
ax1.set_ylabel(r"$f(x)$", fontsize=20)  # Лэйбл по оси Оу
ax1.axvline(x=0, lw=1, color="black")  # Построим х=0
ax1.axhline(y=0, lw=1, color="black")  # Построим у=0
ax1.plot(X, Y, "o--", lw=2, label="Main Function")  # График подпоследовательнсоти
ax1.plot(X, Y_1, label="Polynom: n = 1")
ax1.plot(X, Y_2, label="Polynom: n = 2")
ax1.plot(X, Y_9, label="Polynom: n = 9")
ax1.legend(loc="center left", bbox_to_anchor=(1, 0.5), edgecolor="black")
```
