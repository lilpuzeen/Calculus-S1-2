```python
A = np.array([[0, 3, 0], 
              [-3, -3, 0], 
              [0, -3, -3]])
E = np.array([[1, -1, 1], 
              [-1, 2, -3], 
              [3, -4, 6]])
H = np.array([[-1, -2, -2], 
              [2, 3, 6], 
              [-2, -2, -7]])
E_new = np.array([[-1, 2, 1], 
                  [1, -3, -3], 
                  [-2, 5, 5]])
H_new = np.array([[-1, 2, -2], 
              [1, -1, 4], 
              [-1, 3, 1]])
# B = T_H^-1 dot A dot T_E
T_H_Inv = np.linalg.inv(H_new).dot(H)
T_E = np.linalg.inv(E).dot(E_new)
np.round(T_H_Inv.dot(A).dot(T_E))
```

<p style="font-size: 20px">В ответ вам нужно записать (см. формат ответа в геолине): <br>собственное число [собственный(ые) вектор(а), ...; присоединенный (ые) вектор(а), ...]</p>

<p style="font-size: 20px">Для данной задачи в ответ пойдет</p> <br>
-2 [-1, -1, 1; -1, -2, 0] <br> 2 [2, 3, 0]

### Чтобы удостовериться в ответе, найдите нормальную Жорданову форму матрицы

```python
m = Matrix([[11, -6, 7], [21, -12, 11], [3, -2, -1]])
P, J = m.jordan_form()
J
```

<p style="font-size: 20px">Можем видеть что у нас есть блок 2х2 для $\lambda = -2$ и блок 1x1 $\lambda = 2$</p>
<p style="font-size: 20px">Следует что у $\lambda = -2$ есть собственный вектор и один присоединенный (для присоединенных смотрим на единички над главной диагональю, если есть, значит это присоединенный)</p>

<p style="font-size: 20px">Но у $\lambda = 2$ нет следующих столбцов с 1 над ГД $\rightarrow$ у этого С.З. не будет присоединенных векторов</p>
