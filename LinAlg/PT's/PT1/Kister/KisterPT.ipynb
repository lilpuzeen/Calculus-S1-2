```python
def to_string(a):
	result = ""
	for i in range(len(a)):
		for j in range(len(a[i])):
			if j != len(a) - 1:
				result += str(a[i][j]) + ", "
			else:
				result += str(a[i][j]) + "; "
	return "[" + result[:-2] + "]"
```

```python
A = np.array([[13, 4, 2, 1], 
              [5, 8, 2, 1], 
              [-15, -8, 0, -1], 
              [-39, -20, -10, -3]])
to_string(2*sp.cosm(A) + sp.sqrtm(A))
```
