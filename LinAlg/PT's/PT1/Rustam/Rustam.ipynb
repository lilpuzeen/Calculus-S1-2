```python
def to_string(a):
	result = ""
	for i in range(len(a)):
		for j in range(len(a[i])):
			if j != len(a) - 1:
				result += str(a[i][j]) + ", "
			else:
				result += str(a[i][j]) + "; "
	return "[" + result[:-2] + "]"
```

```python
A = np.array([[2, 1, 0, 0], [5, 3, 2, 1], [-7, -4, -1, -1], [-7, -4, -2, 0]])
Asin = sp.sinm(A)
res = 3 * (Asin @ Asin)
to_string(np.round(res, 6))
```
