```python
def to_string(a):
	result = ""
	for i in range(len(a)):
		for j in range(len(a[i])):
			if j != len(a) - 1:
				result += str(a[i][j]) + ", "
			else:
				result += str(a[i][j]) + "; "
	return "[" + result[:-2] + "]"
```

```python
A = np.array([[10, 4, 2, 1], [-7, -1, -2, -1], [-19, -10, -2, -4], [17, 8, 4, 8]])
Asin = sp.sinm(A)
res = 3 * (Asin @ Asin)
to_string(np.round(res, 7))
```
