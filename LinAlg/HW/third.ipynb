```python
A = np.array([[3, -1, -3, 4], 
              [-1, 1, 0, 0], 
              [-1, 0, 1, -1]])
E = np.array([[-1, -1, 1, -1], 
              [2, 3, -4, 0], 
              [-3, -4, 6, -3], 
              [-6, -8, 11, -3]])
H = np.array([[1, -2, 1], 
              [1, -1, 2], 
              [2, -3, 4]])
E_new = np.array([[1, 1, 1, 1], 
                  [1, 2, 2, -1], 
                  [-3, -4, -3, 0], 
                  [8, 11, 9, 1]])
H_new = np.array([[1, -2, -1], 
              [-2, 5, 0], 
              [5, -12, 0]])
# B = T_H^-1 dot A dot T_E
T_H_Inv = np.linalg.inv(H_new).dot(H)
T_E = np.linalg.inv(E).dot(E_new)
np.round(T_H_Inv.dot(A).dot(T_E))
```

```python
A = np.array([[3, 3, 3], 
              [3, 6, 6], 
              [-3, -6, -3]])
E = np.array([[-1, 1, -2], 
              [2, -1, 5], 
              [-3, 2, -6]])
E_new = np.array([[1, -1, -1], 
                  [-2, 3, 3], 
                  [3, -4, -3]])
# T = E^-1 dot ~E
T = np.round(np.linalg.inv(E).dot(E_new))
S = np.linalg.inv(T)
S.dot(A).dot(T)
```
