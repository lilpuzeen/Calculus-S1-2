```python
import numpy as np
import numpy.linalg as la
```

```python
def fast_reshape_row_to_col(row_vector):
    coordinates = row_vector.shape
    return row_vector.reshape((coordinates[1], 1))
```

```python
def fast_reshape_col_to_row(col_vector):
    coordinates = col_vector.shape
    return col_vector.reshape(())
```
