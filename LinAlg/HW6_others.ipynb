```python
import numpy as np
import numpy.linalg as la

a_4 = np.array([[[0, -5], [0, 3]], [[3, 5], [-5, -4]]])

e_4 = np.array([[1, 1], [-2, -1]])

e1_4 = np.array([[-1, -1], [-1, 0]])

t_4 = la.inv(e_4) @ e1_4
s_4 = la.inv(t_4)

print(np.einsum('ih, nj, kl, nli -> jkh', t_4, t_4, s_4, a_4))
```
