```python
import numpy as np
from sympy import *
import scipy.linalg as sp
```

```python
A = np.array([[[[3, -6], 
              [2, 0]], 
             [[1, -4], 
              [-2, 3]]], 
            [[[-5, -2], 
              [-5, -2]], 
             [[1, 1], 
              [-2, 4]]]])

v0 = np.array([5, -1])

v1 = np.array([-6, -3])

v2 = np.array([4, -5])

v3 = np.array([5, 2])

c = 0

for p in range (2):
    for l in range (2):
        for m in range (2):
            for i in range (2):
                c += A[i][m][p][l] * v0[p] * v1[l] * v2[m] * v3[i]
                
c
```

```python
A = np.array([[[-1, -5], [2, 4]], [[4, 1], [0, -2]]])

u0 = np.array([-6, 4])

u1 = np.array([5, -2])

v0 = np.array([1, -4])


c = 0;

for j in range (2):
    for n in range (2):
        for l in range (2):
            c += A[l][j][n] * v0[l] * u1[n] * u0[j]
                
c
```

```python
A = np.array([[[2, 0], [-4, -2]], [[-3, -4], [-5, -1]]])

v0 = np.array([1, -1])

v1 = np.array([3, 1])

u0 = np.array([-6, -3])


c = 0;

for p in range (2):
    for t in range (2):
        for n in range (2):
            c += A[n][p][t] * v0[t] * v1[n] * u0[p]
                
c
```
