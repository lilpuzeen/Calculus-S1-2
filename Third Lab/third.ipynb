### Функция и отрезок:
$${\normalsize f(x) = 2^x \in [-1, 2]}$$

<h2 style="text-align:center">Аналитический Метод</h2>

<p style="font-weight: bold; font-size: 18px;">1. Построим верхние и нижние суммы Дарбу для равномерного разбиения</p>

<p style="font-size: 16px;">Пусть:</p>

$${\normalsize \tau - \text{равномерное разбиение отрезка на } n \text{ частей}}$$

<p style="font-size: 16px;">Исходя из того что мы работаем с отрезком $[-1, 2]$ получим:</p>

$${\normalsize \Delta{x_i} = \frac{3}{n} \quad x_i = -1 + \frac{3 i}{n} \quad i = 0, 1, \ldots, n}$$

<p style="font-size: 16px;">Посчитаем нижние и верхние суммы Дарбу:</p>

<p style="font-size: 16px;">Буду пользоваться формулой суммы геометрической прогрессии:</p>

$${\normalsize S_n = \frac{b_1 \cdot (q^n - 1)}{q - 1}}$$

<p style="font-size: 16px; text-align: center">Нижняя сумма Дарбу</p>

$${\normalsize s_{\tau} = \sum_{i = 0}^{n - 1}{\frac{2^{-1 + \frac{3 i}{n}} \cdot 3}{n}} = \frac{3}{2n} \sum_{i = 0}^{n - 1}{\left(8^{\frac{1}{n}}\right)^i} = \frac{21}{2n(\sqrt[n]{8} - 1)}}$$
<hr>

<p style="font-size: 16px; text-align: center">Верхняя сумма Дарбу</p>

$${\normalsize S_{\tau} = \sum_{i = 1}^{n}{\frac{2^{-1 + \frac{3 i}{n}} \cdot 3}{n}} = \frac{3}{2n} \sum_{i = 1}^{n}{\left(8^{\frac{1}{n}}\right)^i} = \frac{21 \cdot \sqrt[n]{8}}{2n(\sqrt[n]{8} - 1)}}$$
<hr>

<p style="font-weight: bold; font-size: 18px;">2. Исследуем интегрируемость функции с помощью критерия Римана</p>

<p style="font-size: 16px; text-align: center">Критерий Римана интегрируемости функции</p>

$${\normalsize \forall \varepsilon,\, \exists \tau: S_{\tau} - s_{\tau} < \varepsilon}$$

<p style="font-size: 16px;">Рассмотрим $S_{\tau} - s_{\tau}$:</p>

$${\normalsize \frac{21 \cdot \sqrt[n]{8}}{2n(\sqrt[n]{8} - 1)} - \frac{21}{2n(\sqrt[n]{8} - 1)} = \frac{21 \cdot \sqrt[n]{8} - 21}{2n(\sqrt[n]{8} - 1)} = \frac{21 \left(\sqrt[n]{8} - 1\right)}{2n(\sqrt[n]{8} - 1)} = \frac{21}{2n}}$$

$${\normalsize S_{\tau} - s_{\tau} = \frac{21}{2n} < \varepsilon}$$

<p style="font-size: 16px;">Найдем такое $n_0$ что $S_{\tau} - s_{\tau} < \varepsilon$:</p>

$${\normalsize \forall \varepsilon, \, \exists n_0: \forall n \ge n_0 \Rightarrow \frac{21}{2n} < \varepsilon \Rightarrow}$$

$${\normalsize \Rightarrow n \gt \frac{21}{2\varepsilon} \Rightarrow n_0 = \left[\frac{21}{2\varepsilon}\right] + 1}$$

<p style="font-size: 16px; text-align: center">Следовательно - функция интегрируема</p>
<hr>

<p style="font-size: 16px; text-align: center">Помимо критерия Римана, интегрируемость можно еще доказать исходя из понятия классов интегрируемых функций</p>

<p style="font-size: 16px; text-align: center">Очевидно что функция $f(x) = 2^x$ монотонно возрастает на отрезке $[-1, 2]$</p>

<p style="font-size: 16px; text-align: center">Это означает что заданная функция принадлежит классу монотонных функций, которые $\in R[a, b]$</p>

<p style="font-weight: bold; font-size: 18px;">3. Найдем пределы сумм Дарбу</p>

<p style="font-size: 16px;">Предел нижней суммы:</p>

$${\normalsize \lim_{n \rightarrow \infty}{s_{\tau}} = \lim_{n \rightarrow \infty}{\frac{21}{2n(\sqrt[n]{8} - 1)}} = \frac{21}{2} \, \lim_{n \rightarrow \infty}{\frac{1}{n \cdot \sqrt[n]{8} - n}} = }$$

<p style="font-size: 16px;">Поделим числитель и знаменатель на $n$ и воспользуемся правилом Лопиталя и получим:</p>

$${\normalsize \frac{21}{2}\lim_{n \rightarrow \infty}{\frac{-\frac{1}{n^2}}{-\frac{\ln{8} \,\cdot\, \sqrt[n]{8}}{n^2}}} = \frac{21}{2 \cdot \ln{8}} \lim_{n \rightarrow \infty}{\frac{1}{8^{\frac{1}{n}}}} = \frac{7}{\ln{4}}}$$

<p style="font-size: 16px;">Предел верхней суммы:</p>

$${\normalsize \lim_{n \rightarrow \infty}{S_{\tau}} = \lim_{n \rightarrow \infty}{\frac{21 \cdot \sqrt[n]{8}}{2n(\sqrt[n]{8} - 1)}} = \frac{21}{2} \, \lim_{n \rightarrow \infty}{\frac{\frac{\sqrt[n]{8}}{n}}{\sqrt[n]{8} - 1}} = }$$

<p style="font-size: 16px;">Воспользуемся правилом Лопиталя и получим:</p>

$${\normalsize \frac{21}{2} \lim_{n \rightarrow \infty}{\frac{\frac{-\ln{8} \,\cdot\, 8^{\frac{1}{n}}\, + \, n \,\cdot\, 8^{\frac{1}{n}}}{n^3}}{-\frac{\ln{8}\, \cdot \,8^{\frac{1}{n}}}{n^2}}} = \frac{21}{2} \lim_{n \rightarrow \infty}{\frac{\ln{8} + n}{n \ln{8}}} = \frac{21}{2 \cdot \ln{8}} \lim_{n \rightarrow \infty}{\frac{\frac{\ln{8}}{n} + 1}{1}} = \frac{7}{\ln{4}}}$$

<p style="font-size: 16px; text-align: center">В пункте 2, мною была доказана интегрируемость нашей функции. Исходя из этого можно утверждать что:</p>

<p style="font-size: 16px; text-align: center">Так как функция монотонно возрастает на отрезке $[-1, 2]$, то:</p>

$${\normalsize I_* = \sup{s_{\tau}} = \frac{7}{\ln{4}} \quad \left(s_{\tau} < \frac{7}{\ln{4}} \quad \text{&&} \quad  \lim_{n \rightarrow \infty}{s_{\tau}} = \frac{7}{\ln{4}} \right)}$$

$${\normalsize I^* = \inf{S_{\tau}} = \frac{7}{\ln{4}} \quad \left(S_{\tau} > \frac{7}{\ln{4}} \quad \text{&&} \quad  \lim_{n \rightarrow \infty}{S_{\tau}} = \frac{7}{\ln{4}} \right)}$$

$${\normalsize \Rightarrow I_* = I^* = I = \frac{7}{\ln{4}}}$$

<p style="font-weight: bold; font-size: 18px;">4. Проверим результат с помощью формулы Ньютона-Лейбница</p>

$${\normalsize \int_{-1}^{2}{2^x}\,dx = \left.\frac{2^x}{\ln{2}}\right|_{-1}^{2} = \frac{2^2}{\ln{2}} - \frac{2^{-1}}{\ln{2}} = \frac{7}{\ln{4}}}$$

<p style="font-size: 16px;">Посчитав значение интеграла с помощью формулы Ньютона-Лейбница, удостоверились в правильности вычислений в пункте 3</p>

<p style="font-weight: bold; font-size: 18px;">1. Напишем программу на языке Python 3 для вычисления интегральных сумм с заданными параметрами</p>

```python
import numpy as np  # Библиотека для работы с массивами чисел
import matplotlib.pyplot as plt # Библиотека для построения графиков
import pandas as pd # Библиотека для работы с датафреймами (таблицы)
import random
%matplotlib inline
```

```python
plt.style.use(["science", "notebook", "grid", "high-vis"])  # Стиль координатной плоскости
```

```python
def f(x):  # Желаемую функцию вводить сюда!
    return x**3 + x**2 + 9*x + 3
```

```python
def compute_integral_sums(function, a, b, N, points_attachment): # Подсчет значения интегральной суммы
    difference = (b - a) / N # Разбиение отрезка на N частей
    x = np.linspace(a, b, N+1) # Вектор который хранит в себе точки разбиения
    
    match points_attachment: # Смотрим какое оснащение выбрал пользователь
        case "left":
            left = x[:-1] # Берем левые точки (исключаем самую правую)
            return np.sum(function(left) * difference)
        case "right":
            right = x[1:] # Берем правые точки (исключаем самую левую)
            return np.sum(function(right) * difference)
        case "mid":
            mid = (x[:-1] + x[1:]) / 2 # Берем концы подотрезков, складываем, делим на 2 чтобы получить середину подотрезка
            return np.sum(function(mid) * difference)
        case "random":
            rand = np.random.uniform(x[:-1], x[1:]) # Функция uniform() генерирует случайное число от 0 до 1. Применяем его на векторы левых точек и правых точек
            # тем самым получая рандомную точку на каждом из подотрезков
            return np.sum(function(rand) * difference)
        case _: # Case default
            print("Please, enter 'points_attachment' option and determine the sum kind")
            print("Choose one of the options: [left, right, mid, random]")
```

```python
compute_integral_sums(f, -3, 5, 100, "mid") # Проверить можно здесь) 
```

<p style="font-weight: bold; font-size: 18px;">2. Визуализируем результат работы программы</p>

```python
def riemann_sums_visualization(function, a, b, N):  # Визуализация интегральных сумм (с разными оснащениями)
    # Подготовка нужных векторов (область определения, значений, разбиения)
    n = N
    x = np.linspace(a, b, N+1)
    y = function(x)

    X = np.linspace(a, b, n*N+1)
    Y = function(X)
    
    # Разобьем на 4 подграфика
    fig, axes = plt.subplots(4, 1, figsize=(15, 20))
    
    # График с оснащением в точках слева
    ax_left = axes[0]
    ax_left.axvline(x=0, lw=1, color="black")  # Построим х=0
    ax_left.axhline(y=0, lw=1, color="black")  # Построим у=0
    ax_left.plot(X, Y, 'b') # График функции
    x_left = x[:-1] # Выбираем соответствующие точки (зависит от вида оснащения отрезков)
    y_left = y[:-1]
    ax_left.plot(x_left, y_left, "b.", markersize=10) # Точки на самом графике функции
    ax_left.bar(x_left, y_left, width=(b - a)/N, alpha=0.2, align="edge", edgecolor="b") # Гистограмма (barplot)  
    ax_left.set_title(f"Левая интегральная сумма, N = {N}", fontsize=25) # Название к графику
    # Аналогично для остальных графиков
    
    # График с оснащением в точках посередине
    ax_mid = axes[1]
    ax_mid.axvline(x=0, lw=1, color="black")
    ax_mid.axhline(y=0, lw=1, color="black")
    ax_mid.plot(X,Y,'b')
    x_mid = (x[:-1] + x[1:])/2
    y_mid = function(x_mid)
    ax_mid.plot(x_mid, y_mid,'b.',markersize=10)
    ax_mid.bar(x_mid, y_mid, width=(b-a)/N, alpha=0.2, edgecolor='b')
    ax_mid.set_title(f'Интегральная сумма с точками посередине, N = {N}', fontsize=25)
    
    # График с оснащением в точках справа
    ax_right = axes[2]
    ax_right.axvline(x=0, lw=1, color="black")
    ax_right.axhline(y=0, lw=1, color="black")
    ax_right.plot(X,Y,'b')
    x_right = x[1:]
    y_right = y[1:]
    ax_right.plot(x_right, y_right,'b.',markersize=10)
    ax_right.bar(x_right, y_right, width=-(b-a)/N, alpha=0.2, align='edge', edgecolor='b')
    ax_right.set_title(f'Правая интегральная сумма, N = {N}', fontsize=25)
    
    # График с оснащением в точках рандомно расположенных
    ax_rand = axes[3]
    ax_rand.axvline(x=0, lw=1, color="black")
    ax_rand.axhline(y=0, lw=1, color="black")
    ax_rand.plot(X,Y,'b')
    x_rand = np.random.uniform(x[:-1], x[1:])
    y_rand = function(x_rand)
    ax_rand.plot(x_rand, y_rand,'b.',markersize=10)
    ax_rand.bar(x_rand, y_rand, width=(b-a)/N, alpha=0.2, align='edge', edgecolor='b')
    ax_rand.set_title(f'Интегральная сумма с рандомными точками, N = {N}', fontsize=25)
    
    fig.tight_layout(h_pad=2.0)
    plt.show()
```

```python
riemann_sums_visualization(f, -3, 5, 10)
```

<p style="font-weight: bold; font-size: 18px;">3. Соберем в таблицу результаты обработки функции и отрезка из аналитической части и сравним их</p>

```python
def compute_error(integral_sum): # Подсчет ошибки относительно получившегося результата в аналитической части и машинным подсчетом
    return np.abs(integral_sum - (7 / np.log(4)))
```

```python
def get_data(): # Собираем значения интегральных сумм для разных разбиений и оснащений
    n_s = np.arange(1000, 105000, 5000)
    points_attachment = ["left", "right", "mid", "random"]
    for i in range(len(n_s)):
        rand_points_attachment = random.choice(points_attachment) # Выбираем рандомное оснащение
        integral_sum = compute_integral_sums(lambda x: 2**x, -1, 2, n_s[i], rand_points_attachment) # Считаем интегральную сумму
        yield ["2^x", n_s[i], rand_points_attachment, integral_sum, compute_error(integral_sum)]
```

```python
information_list = []
for row in get_data():
    information_list.append(row)

# Собираем полученные данные в датафрейм 
df = pd.DataFrame(information_list, columns=["Function", "N", "Points Attachment", "Value Of Riemann Integral Sum", "Calculation Error"])
df
```
